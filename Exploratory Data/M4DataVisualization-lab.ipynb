Import pandas module.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
from matplotlib import pyplot as plt
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY, conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: How to describe a table


```python
table_name = "master"  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(
    table_name
)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0, 0])
```

# Hands-on Lab


## Visualizing distribution of data


```python
# Let get our data before visualizing
QUERY = """
SELECT*
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
plot1 = sns.histplot(df["ConvertedComp"])
plot1.set_xlabel("ConvertedComp")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
plot2 = sns.boxplot(df["Age"])
plot2.set_xlabel("Age")
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

plot3 = sns.scatterplot(x="Age", y="WorkWeekHrs", data=df)
plot3.set_xlabel("Age")
plot3.set_ylabel("WorkWeekHrs")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plot4 = plt.scatter(df["WorkWeekHrs"], df["CodeRevHrs"], alpha=0.5, s=df["Age"] * 10)
plot4
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
Query = """
SELECT * FROM DatabaseDesireNextYear
"""
# Get your data
df_db = pd.read_sql_query(Query, conn)
data_db = (
    df_db["DatabaseDesireNextYear"]
    .value_counts()
    .head()
    .reset_index()
    .rename(columns={"DatabaseDesireNextYear": "Databases", "count": "Count"})
)

# Plot your pie
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

plt.figure(figsize=(15, 6))
plt.pie(
    x=data_db['Count'],
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    labels=None,
    pctdistance=1.12,
)
plt.title('Wishlist of Database', y=1.12) 
plt.axis('equal') 
plt.legend(labels=data_db['Databases'], loc='upper left') 

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df_2 = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
df_2[['WorkWeekHrs','CodeRevHrs']].describe()

df_2.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median().plot.bar(stacked=True)
plt.title('Median hours by Age')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_3 = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
df_3.groupby('Age')['ConvertedComp'].median().plot.line()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
df.groupby('MainBranch')['MainBranch'].count().plot(kind='barh', figsize=(10, 6))
```
